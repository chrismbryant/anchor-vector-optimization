# Anchor Vector Optimization

```python
import os

import numpy as np
from scipy.integrate import trapz
import tensorflow as tf
from tqdm import tqdm

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

anchors = [
    np.array([0.10, 0.20]),
    np.array([0.33, 0.81]),
    np.array([0.67, 0.52]),
    np.array([0.60, 0.84])
]

dist = lambda a, b: np.linalg.norm(a - b)

def f(x, y, anchors, p=2, q=1):
    
    result = 1
    for a in anchors:
        norm = ((x - a[0])**p + (y - a[1])**p)**(1/p)
        result *= norm
    
    result ** 2
    
    return (1 - result)**q

def sigmoid(x, w = 1, b = 0):
    
    s = 1 / (1 + np.exp(-x * w + b))
    return s
```

# Set up Model

```python
def random_mini_batches(X, Y, minibatch_size, seed):
    
    np.random.seed(seed)
    
    _, m = X.shape
    
    indices = np.array([i for i in range(m)])
    
    np.random.shuffle(indices)
    
    X = X[:, indices]
    Y = Y[:, indices]
    
    minibatches = []
    i = 0
    
    while i < m:
        
        minibatches.append((X[:, i:i+minibatch_size], Y[:, i:i+minibatch_size]))
        i += minibatch_size
    
    return minibatches

def create_placeholders(n, k):
    
    X = tf.placeholder(tf.float32, shape = [n, None], name = "X")
    Y = tf.placeholder(tf.float32, shape = [k, None], name = "Y")
    
    return X, Y

def initialize_params(p, n_x):
    
    # p anchors
    # n features
    
    A = tf.get_variable("A", [p, n_x], dtype = tf.float32, initializer = tf.contrib.layers.xavier_initializer(seed = 1))
    w = tf.get_variable("w", [1], dtype = tf.float32, initializer = tf.zeros_initializer())
    b = tf.get_variable("b", [1], dtype = tf.float32, initializer = tf.zeros_initializer())
    
    params = {
        "A": A,
        "w": w,
        "b": b
    }
                        
    return params

def forward_propagate(X, params):

    # A.shape = (p, n)
    # X.shape = (n, m)
    
    A = params["A"]
    w = params["w"]
    b = params["b"]
    
    p = tf.shape(A)[0]
    m = tf.shape(X)[1]
    
    norm_A = tf.reshape(tf.norm(A, axis = 1), (p, 1))
    norm_X = tf.reshape(tf.norm(X, axis = 0), (1, m))
    norm = tf.matmul(norm_A, norm_X)
    
    sim = tf.divide(tf.matmul(A, X), norm) # shape = (p, m)
    dist = 0.5*(1 - sim)
    
    D = tf.reduce_prod(dist, axis = 0) # shape = (1, m)
    H = tf.sigmoid(w * D + b)
    
    return H

def get_cost(Y, H, epsilon = 0.0001):
    
    # Y.shape = (1, m)
    # H.shape = (1, m)
    
    m = tf.cast(tf.shape(Y)[1], "float32")
    cost = - 1 / m * tf.reduce_sum((Y * tf.log(tf.maximum(H, epsilon)) + (1 - Y) * tf.log(tf.maximum(1 - H, epsilon))))
    
    return cost

def get_A_reg(A, q=0, epsilon = 0.0001):
    
    # A.shape = (p, n)
    
    if q == 0:
        return 0
    
    p = tf.shape(A)[0]
    
    norm_A = tf.reshape(tf.norm(A, axis = 1), (p, 1))
    norm = tf.matmul(norm_A, tf.transpose(norm_A))
    
    sim = tf.divide(tf.matmul(A, tf.transpose(A)), norm)
    dist = tf.maximum(1 - sim, epsilon)
    
    energy = 1/2 * tf.reduce_sum(q**2 / dist)
    
    return energy

##### -------------------------------------------------------- #####

def model(X_train, 
          Y_train, 
          X_test, 
          Y_test, 
          p = 4,
          q = 0,
          learning_rate = 0.0001,
          num_epochs = 1500, 
          minibatch_size = 32, 
          print_cost = True,
          keep_param_history = False,
          img_path=None):
    
    # X_train.shape = (n_x, m)
    # Y_train.shape = (1, m)
    
    tf.reset_default_graph()
    
    tf.set_random_seed(1)
    seed = 2
    
    (n_x, m) = X_train.shape
    n_y = Y_train.shape[0]
    J_train = []
    J_test = []
    
    X, Y = create_placeholders(n_x, n_y)
    params = initialize_params(p, n_x)
    H = forward_propagate(X, params)
    J = get_cost(Y, H) + get_A_reg(params["A"], q = q)
    
    optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(J)

    init = tf.global_variables_initializer()
    
    # Initialize list of parameters
    param_history = []
    
    with tf.Session() as sess:
        
        sess.run(init)
        
        for epoch in tqdm(range(num_epochs)):

            epoch_cost = 0
            num_minibatches = int(m / minibatch_size)
            seed += 1
            minibatches = random_mini_batches(X_train, Y_train, minibatch_size, seed)
            
            for minibatch in minibatches:

                (minibatch_X, minibatch_Y) = minibatch
                _ , minibatch_cost = sess.run([optimizer, J], feed_dict={X: minibatch_X, Y: minibatch_Y})
                epoch_cost += minibatch_cost / num_minibatches
            
            if print_cost == True and epoch % 200 == 0:
                print ("Cost after epoch %i: %f" % (epoch, epoch_cost))
                
            if epoch % 5 == 0:
                J_train.append(epoch_cost)
                J_test.append(sess.run(J, feed_dict = {X: X_test, Y: Y_test}))
            
            def append_to_params():
                param_history.append({"epoch": epoch, "params": sess.run(params)})
                return
            
            if epoch < 10:
                append_to_params()
            elif epoch < 20 and epoch % 2 == 0:
                append_to_params()
            elif epoch <= 50 and epoch % 5 == 0:
                append_to_params()
            elif epoch % 20 == 0:
                append_to_params()

        params = sess.run(params)
        probs_train = sess.run(H, feed_dict = {X: X_train})
        probs_test = sess.run(H, feed_dict = {X: X_test})
        
        print("Training complete.")
        
        result = {
            "probs_train": probs_train,
            "probs_test": probs_test,
            "J_train": np.squeeze(J_train),
            "J_test": np.squeeze(J_test),
            "params": params
        }
        
        plot_J(
            result["J_train"],
            result["J_test"],
            p,
            q,
            learning_rate,
            img_path=img_path
        )
        
        if keep_param_history:
            result["param_history"] = param_history
        
        return result
    
def plot_J(J_train, J_test, p, q, learning_rate, img_path=None):
    
    fig = plt.figure(figsize = (10, 7))
    ax = fig.add_subplot(1, 1, 1)
    
    font = {"family": "DejaVu Sans", "weight": "normal", "size" : 14}
    plt.rc("font", **font)
    
    ax.plot(J_train, label = "J_train")
    ax.plot(J_test, label = "J_test")
    ax.legend()
    ax.set_xlim([0, len(J_train)])
    ax.set_ylabel("Cost", fontsize = 16)
    ax.set_xlabel("Epoch/5", fontsize = 16)
    ax.set_title("%d Anchors: q = %g,  = %g" % (p, q, learning_rate), fontsize=20)
    
    if img_path:
        plt.savefig(img_path, dpi=300, bbox_inches="tight")
    
    plt.show()
    plt.gcf().clear()
    plt.close()
    
    return
```

 # Define evaluation functions

```python
def compute_roc(y, probs, N=500):
    """
    Compute the TPR and FPR values for a series of threshold values, as well
    as the AUC (area under the curve) of the resulting ROC (receiver operating
    characteristic) curve, given an input list of labels and prediction
    probabilities.

    PARAMETERS
    ----------
    y -- list or Numpy array of binary labels (0 or 1).
    
    probs -- list or Numpy array of prediction probabilities.

    N -- number of threshold values (default: 500).
    RETURNS
    -------
    roc -- dictionary of ROC-related values:
      {
        "thresh": numpy array of decision threshold values between 0 and 1,
        "tpr": numpy array of TPR values (1 for each threshold value),
        "fpr": numpy array of FPR values (1 for each threshold value),
        "auc": area under the ROC curve
      }.
    """
    thresh_values = np.arange(N)/N

    probs = np.array(probs)
    y = np.array(y)

    pos = np.sum(y == 1)
    neg = np.sum(y == 0)

    roc_tpr = []
    roc_fpr = []

    for thresh in thresh_values:

        predictions = (probs > thresh).astype(int)
        tp = np.sum(np.logical_and(predictions == 1, y == 1))
        fp = np.sum(np.logical_and(predictions == 1, y == 0))
        tpr = tp/pos
        fpr = fp/neg
        roc_tpr.append(tpr)
        roc_fpr.append(fpr)

        auc = -trapz(roc_tpr, roc_fpr)

        roc = {
        "tpr" : roc_tpr,
        "fpr" : roc_fpr,
        "auc" : auc,
        "thresh": thresh_values
        }

    return roc

def plot_roc(y, probs, title="ROC Curve", cmap="rainbow", N=500, img_path=None):
    """
    Compute and plot the ROC curve given an input list of labels
    and prediction probabilities.

    PARAMETERS
    ----------
    y -- list or Numpy array of binary labels (0 or 1).
    
    probs -- list or Numpy array of prediction probabilities.

    title -- title of plot (default: "ROC Curve").

    cmap -- colormap string specification (default: "rainbow")

    N -- number of threshold values (default: 500).

    img_path -- path to save image (default: None).

    RETURNS
    -------
    roc -- dictionary of ROC-related values:
      {
        "thresh": numpy array of decision threshold values between 0 and 1,
        "tpr": numpy array of TPR values (1 for each threshold value),
        "fpr": numpy array of FPR values (1 for each threshold value),
        "auc": area under the ROC curve
      }.
    """

    roc = compute_roc(y, probs, N)
    thresh_values = roc["thresh"]

    font = {"family": "DejaVu Sans", "weight": "normal", "size" : 14}
    plt.rc("font", **font)
    
    fig = plt.figure(figsize = (10, 7))
    ax = fig.add_subplot(1, 1, 1, aspect = "equal")
    ax.grid(True)
    
    ax.scatter(roc["fpr"],
               roc["tpr"],
               s = 100,
               c = thresh_values,
               cmap = cmap,
               marker = ".",
               edgecolors = "none")

    # Plot line of randomness
    ax.plot([0, 1], [0, 1], "k-")
    
    # Make Color Bar
    norm = matplotlib.colors.Normalize(0.0, 1.0)
    sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
    sm.set_array([])
    cbar = fig.colorbar(sm, ticks=np.linspace(0, 1, 11))
    cbar.set_label("Threshold Value")

    ax.text(x = 0.7,
            y = 0.1,
            s = "AUC = %.3f" % roc["auc"],
            ha = "left",
            va = "center",
            bbox=dict(facecolor="gray",
                      alpha=0.1,
                      boxstyle="round"))

    ax.set_xlabel("FPR [= FP/(FP + TN)]")
    ax.set_ylabel("TPR [= TP/(TP + FN)]")
    ax.set_title(title, fontsize = 20)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    
    if img_path:
        plt.savefig(img_path, dpi=300, bbox_inches="tight")
    
    plt.show()
    plt.gcf().clear()
    plt.close()

    return roc
```

# Run Model

Charged anchor optimization: Add charge as a regularization parameter enforcing that all the anchors are as spread apart as possible while still minimizing the loss. (i.e. introduce a tunable electrostatic energy minimization condition to the total loss function.)

* U = sum(q^2/R_ij) over i and j,
* R_ij = cosine distance between anchor i and j where i != j,
* q = anchor charge

```python
p = 3
q = 0
alpha = 0.01
num_epochs = 5000

result = model(X_train, 
               Y_train, 
               X_test, 
               Y_test, 
               p = p,
               q = q,
               learning_rate = alpha,
               num_epochs = num_epochs,
               print_cost = False,
               keep_param_history = True,
               img_path="images/anchor_training/cost.png")

probs_train = result["probs_train"]
probs_test = result["probs_test"]
params = result["params"]
param_history = result["param_history"]
```

# Plot learned probability mapping

```python
def plot_probability_map(params, img_path=None, epoch=None, dpi=None, show=True):

    def get_ang_cos_sim(A_phi, A_theta, phi, theta):

        sim = []

        for i in range(len(A_phi)):

            s = np.sin(A_theta[i]) \
                * np.dot(np.atleast_2d(np.cos(A_phi[i] - phi)).T, np.atleast_2d(np.sin(theta))) \
                + np.cos(A_theta[i]) * np.cos(np.atleast_2d(theta))

            sim.append(s)

        return np.array(sim)

    def get_D(dist):

        return np.prod(dist, axis = 0)


    phi = np.linspace(-np.pi, np.pi, 100)
    theta = np.linspace(0, np.pi, 100)
    Phi, Theta = np.meshgrid(phi, theta)

    A_learned = params["A"]
    A_r, A_theta, A_phi = cart_to_sphere(A_learned[:, 0], A_learned[:, 1], A_learned[:, 2])
    sim = get_ang_cos_sim(A_phi, A_theta, phi, theta)
    dist = 0.5*(1 - sim)
    D = get_D(dist) ** 1
    D = sigmoid(params["w"] * D + params["b"])

    fig = plt.figure(figsize = (15, 10))
    ax = fig.add_subplot(1, 1, 1)
    ax.contourf(Phi.T, Theta.T, D, 1000, cmap = "coolwarm")
    ax.scatter(phi_pos, theta_pos, c = "green", s = 30)
    ax.scatter(phi_neg, theta_neg, c = "orange", s = 30)
    ax.scatter(A_phi, A_theta, c = "yellow", marker = 'x', s = 20, linewidth = 20)
    ax.axis("off")
    
    if epoch is not None:
        ax.text(x = -3,
            y = 0.1,
            s = "Epoch = %d" % epoch,
            ha = "left",
            va = "center",
            bbox=dict(facecolor="lightgray",
                      alpha=0.8,
                      boxstyle="round"))
    
    if img_path:
        dpi = 60 if dpi is None else dpi
        plt.savefig(img_path, dpi=dpi, transparent=True, bbox_inches="tight")
        
    if show:
        plt.show()
        
    plt.gcf().clear()
    plt.close()
    
    return
```
